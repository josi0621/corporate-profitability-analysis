```python
cols = [
    'ROA - Return On Assets',
    'Net Profit Margin',
    'Debt/Equity Ratio',
    'Gross Margin',
    'Operating Margin',
    'Current Ratio'
    ]
df_subset = df[cols].copy()
df_subset.isnull().sum()
```

```python
df_subset.head()
```

```python
df_subset.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(df_subset.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Profitability and Financial Ratios")
plt.show()
```

```python
sns.scatterplot(data=df_subset, x='Net Profit Margin', y='ROA - Return On Assets')
plt.title("ROA vs Net Profit Margin")
plt.xlabel("Net Profit Margin")
plt.ylabel("Return On Assets (ROA)")
plt.show()
```

```python
sns.scatterplot(data=df_subset, x='Operating Margin', y='ROA - Return On Assets')
plt.title("ROA vs Operating Margin")
plt.xlabel("Operating Margin")
plt.ylabel("Return On Assests (ROA)")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

x = df_subset[['Net Profit Margin', 'Operating Margin', 'Debt/Equity Ratio', 'Gross Margin', 'Current Ratio']]
y = df_subset['ROA - Return On Assets']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print("R-squared: ", r2)
print("Mean Squared Error: ", mse)
```

```python
coefficients = pd.DataFrame({
    'Feature': x.columns,
    'Coefficient': model.coef_
})

print(coefficients)
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.xlabel("Actual ROA")
plt.ylabel("Predicted ROA")
plt.title("Actual vs Predicted ROA")
plt.plot([y.min(), y.max()], [y.min(), y.max()], '--r')
plt.show()
```

```python
x_npm = df_subset[[
    'ROA - Return On Assets',
    'Debt/Equity Ratio',
    'Gross Margin',
    'Operating Margin',
    'Current Ratio'
    ]]
y_npm = df_subset['Net Profit Margin']
```

```python
model_npm = LinearRegression()
model_npm.fit(x_train_npm, y_train_npm)
```

```python
y_pred_npm = model_npm.predict(x_test_npm)
r2_npm = r2_score(y_test_npm, y_pred_npm)
mse_npm = mean_squared_error(y_test_npm, y_pred_npm)

print("R-squared: ", r2_npm)
print("Mean Squared Error: ", mse_npm)
```

```python
coefficients_npm = pd.DataFrame({
    'Feature': x_npm.columns,
    'Coefficient': model_npm.coef_
})

print(coefficients_npm)
```
